# 5 Number Summary and Box Plot

```python
import numpy as np
```

```python
marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
```

```python
mini,Q1,median,Q3,max = np.quantile(marks,[0,0.25,0.50,0.75,1.0])
```

```python
mini,Q1,median,Q3,max
```

```python
iqr = Q3 - Q1
```

```python
lower_fence = Q1 - 1.5*iqr
lower_fence
```

```python
higher_fence = Q3 + 1.5 * iqr
higher_fence
```

```python
import seaborn as sns
```

```python
sns.boxplot(marks,orient='h')
```
