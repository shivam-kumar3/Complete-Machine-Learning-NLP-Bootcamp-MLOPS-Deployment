Norminal/ OHE Encoding 
One hot encoding, also known as nominal encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algo. 
In this technique, each category is represented as binary vector where each bit corresponds to be unique category.
eg. if we have a categorical variable "color" with three possible values (red,green,blue) we can represent it in one hot encoding as follow

1. Red(1,0,0)
2. Green(0,1,0)
3. Blue (0,0,1)

OHE will create the numbers of features in the dataset

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
# Creating a simple dataframe

df = pd.DataFrame({
    'color' : ["red", "blue","green","green","red","blue"]
})
```

```python
df.head()
```

```python
# create an instance of Onehotencoder
encoder = OneHotEncoder()
```

```python
# perform fit and then transform

encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
df_final = pd.concat([df,encoded_df],axis=1)
```

```python
df_final
```
