SMOTE

* SMOTE (Synthetic Minority Over-sampling Technique) is a method used to address class imbalance in datasets by generating synthetic samples of the minority class. Instead of simply duplicating existing minority instances, SMOTE creates new samples by interpolating between existing minority class examples and their nearest neighbors. This helps improve the performance of machine learning models on underrepresented classes.





```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                    weights=[0.90],random_state=12)
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
```

```python
df = pd.concat([df1,df2],axis=1)
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['f1'],df['f2'],c = df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# transform the data
oversample = SMOTE()
x,y = oversample.fit_resample(df[['f1','f2']],df['target'])
```
