MATPLOTLIB

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [5,6,8,10,15]
```

```python
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.title("Basic line plot")
```

CUSTOMIZED LINE PLOT

```python
x = [1,2,3,4,5]
y = [5,6,8,10,15]

plt.plot(x,y, color = "red", linestyle = "--",linewidth = 3, marker = "*", markersize = 10 ,)
plt.grid(True)
```

Multiplots

```python
x = [1,2,3,4,5]
y = [1,4,7,8,10]
y1 = [5,4,32,12,13]
lang = ["Python", "C++", "C", 'Java', "Java Script"]
demand = [50,10,20,30,15]

plt.figure(figsize=(10,4))
plt.subplot(1,5,1)
plt.scatter(x,y)

plt.subplot(1,5,2)
plt.plot(y,y1)

plt.subplot(1,5,3)
plt.bar(x,y)

plt.subplot(1,5,4)
plt.pie(demand,labels=lang)

plt.subplot(1,5,5)
plt.hist(demand, bins = 5, edgecolor = "black", color = "yellow")
```

```python
# scatter plot
x = [1,2,3,4,5]
y = [5,6,8,10,15]

plt.scatter(x,y)
```

```python
# pie chart

lang = ["Python", "C++", "C", 'Java', "Java Script"]
demand = [50,10,20,30,15]
explode = [0.2,0.2,0.2,0.2,0]

plt.pie(demand, labels= lang, colors = ["red",'green','pink','yellow','blue'], explode=explode, autopct = "%1i%%")
plt.show()
```
