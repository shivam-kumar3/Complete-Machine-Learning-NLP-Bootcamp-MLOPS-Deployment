Data Visualization with seaborn

Seaborn is a python visulization library based on matplotlib that provides a high level interface for drawing attractive and informative statistical graphics.
Seaborn helps in creating complex visulization with just a few lines of code.
In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Basic plotting with seaborn
tips = sns.load_dataset('tips')

```

```python
tips
```

```python
sns.scatterplot(x = 'sex', y= 'tip', data = tips)
plt.title("Testing")
plt.show()
```

```python
# Line plot

sns.lineplot(x ='sex' ,y = 'tip', data = tips)
```

```python
## categorical plot

# barplot

sns.barplot(x = 'total_bill', y='tip', data = tips)
```

```python
# boxplot

sns.boxplot(x= 'day',y='total_bill', data = tips)
```

```python
# violin plot

sns.violinplot(x= 'day',y='total_bill', data = tips)
```

```python
# histograms

sns.histplot(tips['total_bill'], bins = 10, kde = True)
```

```python
# kde plot

sns.kdeplot(tips['total_bill'], fill= True)
```

```python
# pairplot

sns.pairplot(tips)
```

```python
# heatmap
cor = tips[['tip','total_bill','size']].corr()
sns.heatmap(cor, cmap = 'coolwarm')
```
