```python
# Array initializatin methods
```

```python
# zero array
zero_arr = np.zeros((4,5))
print(zero_arr)
```

```python
# once array
one_array = np.ones((4,5))
print(one_array)
```

```python
# full array
full_array = np.full((4,5),7)
print(full_array)
```

```python
# Identity Matrix
id_arr = np.eye(3)
print(id_arr)
```

```python
# Empty array
np.empty(2)
```

```python
# evenly spaced array
np.arange(1,10,2)
```

```python
# specific numbers of values equally spaced bw a range 
np.linspace(1,10,4)
```

```python
# random values array - float
r_array = np.random.rand(3,2)
print(r_array)
```

```python
# random value array - int
r_arr = np.random.randint(2,8,(3,3))
print(r_arr)
```
