```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logi = LogisticRegression()
logi.fit(x_train,y_train)
```

```python
y_pred = logi.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred)*100
print(score)

cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Example: Create dummy data
X = np.random.rand(100, 10)  # 100 samples, 10 features
y = np.random.randint(0, 2, size=100)  # Binary target variable

# Check shapes of X and y
print("X shape:", X.shape)  # Should be (n_samples, n_features)
print("y shape:", y.shape)  # Should be (n_samples,)

# Initialize the logistic regression model
logi = LogisticRegression(max_iter=1000)  # Ensure convergence

# Define the hyperparameters to search
penalties = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 1.0, 0.1, 0.01]
solvers = ['liblinear', 'saga']  # Include 'saga' for elasticnet

# Create a parameter grid (note: scoring is not included here)
params = {
    'penalty': penalties,
    'C': c_values,
    'solver': solvers
}

# Setup stratified k-fold cross-validation (use StratifiedKFold if not using groups)
cv = StratifiedKFold(n_splits=5)

# Initialize GridSearchCV (move scoring out of param_grid)
grid = GridSearchCV(
    estimator=logi,
    param_grid=params,
    scoring='accuracy',  # Set scoring directly
    cv=cv,
    n_jobs=-1,
    refit=True  # Optionally refit the model with the best parameters
)

# Fit the grid search
grid.fit(X, y)

# To access the best parameters and the best estimator
best_params = grid.best_params_
best_estimator = grid.best_estimator_

print("Best parameters:", best_params)
print("Best estimator:", best_estimator)
```
