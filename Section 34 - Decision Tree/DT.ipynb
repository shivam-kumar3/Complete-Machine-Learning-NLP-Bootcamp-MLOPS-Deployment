# Decision Tree

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
df
```

```python
columns = [['sepal length', 'sepal width', 'petal length', 'petal width']]
x = pd.DataFrame(df['data'], columns=columns)
y = df['target']
```

```python
x
```

```python
y
```

```python
# train test split 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier = DecisionTreeClassifier()
treeclassifier.fit(x_train,y_train)
```

```python
# visualization of decision tree

from sklearn import tree
plt.figure(figsize=(10,5))
tree.plot_tree(treeclassifier, filled=True)

```

preventing overfitting 
1 - post pruning 
2 - pre pruning 

```python
treeclassifier = DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(x_train,y_train)
tree.plot_tree(treeclassifier, filled=True)
```

```python
y_pred = treeclassifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

* post pruning - small datasets
    The tree is allowed to grow fully, and then nodes are removed (pruned) based on their contribution to accuracy on a validation set.
* pre pruning - big datasets
    he tree is stopped from growing when further splitting does not provide a significant improvement in the prediction accuracy. This can be based on various criteria

#### Decision Tree Classifier Preprunning and HyperParameter tunning

```python
params= {
    "criterion" : ['gini', 'entropy', 'log_loss'],
    "splitter" :['best','random'],
    "max_depth" : [1,2,3,4,5],
    "max_features" : ['auto','sqrt','log2']
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
tree = DecisionTreeClassifier()
grid = GridSearchCV(tree,param_grid=params,cv = 5, scoring='accuracy')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python
print(confusion_matrix(y_pred , y_test))
print(classification_report(y_pred,y_test))
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print(score)
```
