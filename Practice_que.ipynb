```python
# Write a function to check prime number
def Prime_time(num):
    if num <=1:
        print("Number is not prime")
        return

    for i in range(2,num):
        if num % i ==0:
            print("Number is not prime")
            return
    print('Prime')

Prime_time(7)
```

```python
# Write a function to return the square of all numbers in a list

def square(list):
    lstt = []
    for i in list:
        lstt.append(i*i)
    print(lstt)

lstt = [2,3,4,5,6,7,8]
square(lstt)
```

```python
# Find common elements in two lists

list_1 = [1,2,3,4,5]
list_2 = [4,5,6,7,8]

# convert to set
set_1 = set(list_1)
set_2 = set(list_2)

# common elements
common_elements = set_1 & set_2

print('List 1 :',list_1)
print('List_2 :',list_2)

print('Common elements :',common_elements)
```

```python
# Count frequency of each element in a list

def Frequency_count(input_list):
    frequency = {}
    for i in input_list:
        if i in frequency:
            frequency[i] += 1
        else:
            frequency[i] = 1
    return frequency

listt = [1,1,1,2,3,3,3,4,5,5,5,6,7,7,7,10]

print(Frequency_count(listt))



```
