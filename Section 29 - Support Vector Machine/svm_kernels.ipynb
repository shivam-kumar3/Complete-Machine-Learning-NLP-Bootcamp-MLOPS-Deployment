## SVM KERNELS IMPLEMENTATION

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(-5.0,5.0,100)
y = np.sqrt(10**2 - x**2)
y = np.hstack([y,-y])
x = np.hstack([x,-x])
```

```python
plt.scatter(x,y)
```

```python
x1 = np.linspace(-5.0,5.0,100)
y1 = np.sqrt(5**2 - x1**2)
y1 = np.hstack([y1,-y1])
x1 = np.hstack([x1,-x1])
```

```python
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
import pandas as pd 
df1 = pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df2 = pd.DataFrame(np.vstack([y1,x1]).T,)
```
