```python
from sklearn.datasets import fetch_california_housing
```

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cali = fetch_california_housing()
```

```python
cali
```

```python
df = pd.DataFrame(data= cali.data , columns=cali.feature_names)
```

```python
df.head()
```

```python
df['Price'] = cali.target
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot = True)
```

```python
# Independent and Dependent features
x = df.drop('Price', axis=1) #independent features
y = df['Price'] # dependent features
```

```python
# split the data into train test split 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=10)
```

```python
x_train.shape,
```

```python
# Standardization the data 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
# model training 

from sklearn.linear_model import LinearRegression
regression =LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
# slope or coeff
regression.coef_
```

```python
# Intercept
regression.intercept_
```

```python
# prediction for the test data 

y_pred = regression.predict(x_test)
```

```python
y_pred
```

```python
# R Square and Adjusted R square
from sklearn.metrics import r2_score 
score = r2_score(y_test,y_pred)
print(score)
```

# Pickling

Python pickle module is used for serialising and de- serialising a python object structure. Any object in Python can be pickled so that it can be saved on disk. what pickle does is that it serialise the object first before writing it to file. pickling is a way to convert a python object (list, dict etc ) into a character stream. The idea is that this character streams contains all the information necessart to reconstruct the object in another python script

```python
import pickle
pickle.dump(regression,open('regression.pkl','wb'))
```

```python
model = pickle.load(open('regression.pkl','rb'))
```

```python
model.predict(x_test)
```
