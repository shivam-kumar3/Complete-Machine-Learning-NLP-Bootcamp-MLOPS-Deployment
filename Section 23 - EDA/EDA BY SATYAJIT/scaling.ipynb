# What is Normalization

Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1.

it is also knows as MinMaxScaling

It is generally used when the distribution of the data is not normally distributed

```python
df.describe().round(2)
```

```python
new_df = df[['Age', 'Tenure']]
```

```python
new_df.head()
```

```python
new_df.isnull().sum()
```

```python
scaler = MinMaxScaler()
normalized_df = scaler.fit_transform(new_df)
print(normalized_df)
```

```python
normal = pd.DataFrame(normalized_df,columns=new_df.columns)
```

Standardization
 
This function is used when the data is normally distributed this convert the data where the mean will be 0 and standard deviation will be 1
